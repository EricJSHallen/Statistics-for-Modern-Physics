```python
#importing packages
import numpy as np
import matplotlib.pyplot as plt
import random

#def global var
trials = range(10000)
#number of doors
n = 3

switch_points = 0
random_points = 0
stay_points = 0

for i in trials: 
    doors = np.linspace(1,n,n)
    c_index,car = random.choice(list(enumerate(doors)))
    t_index,contestant = random.choice(list(enumerate(doors)))

    #ind to excl for host choice
    host_excludes = {c_index, t_index}
    #allowed host ind
    h_allowed = [k for k in range(len(doors)) if k not in host_excludes]
    #host ind and door num
    h_index = random.choice(h_allowed)
    host = doors[h_index]

    #switch choice
    switch_excludes = {h_index, t_index}
    switch_allowed = [k for k in range(len(doors)) if k not in switch_excludes]
    #switch ind and door num
    switch_index = random.choice(switch_allowed)
    switch = doors[switch_index]

    #random ind and door num
    rand_excludes = {h_index}
    rand_allowed = [k for k in range(len(doors)) if k not in rand_excludes]
    #random ind and door num
    rand_index = random.choice(rand_allowed)
    rand = doors[rand_index]

    if doors[c_index] == doors[t_index]:
        stay_points += 1
    
    if doors[c_index] == doors[switch_index]:
        switch_points += 1

    if doors[c_index] == doors[rand_index]:
        random_points += 1

print(f"number of trials = {len(trials)}, number of doors = {n} stay {stay_points}, switch {switch_points}, random {random_points}")
    




    
```

```python
categories = ["Switch", "Stay", "Random"]
values     = [switch_points,   stay_points,   random_points]

fig, ax = plt.subplots()
ax.bar(categories, values)  # categories can be strings

ax.set_title(f"points per strategy with number of trials = {len(trials)}, number of doors = {n}")
ax.set_xlabel("strategy")
ax.set_ylabel("counts")

plt.tight_layout()
plt.show()
```
